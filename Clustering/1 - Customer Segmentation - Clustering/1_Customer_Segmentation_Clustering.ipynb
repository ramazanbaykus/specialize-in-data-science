## 2. Veriyi Okuma ve İnceleme

```python
file_path = '/content/customer_segmentation.csv'

df = pd.read_csv(file_path)
df.head()
```

## 3. Veri Temizleme

```python
# Bu bölümde boş veri kontrolü yapacağım ve Income sütunundaki boş değerleri median ile dolduracağım.
```

```python
df.isnull().sum()
```

```python
df['Income'] = df['Income'].fillna(df['Income'].median())
df.isnull().sum()
```

## 4. Feature Engineering

```python
# Bu bölümde ID sütununu çıkaracağım, yaşı hesaplayacağım ve kategorik sütunları sayısal hale getireceğim.
```

```python
df = df.drop('ID', axis=1)
df['Age'] = 2026 - df['Year_Birth']
df = df.drop('Year_Birth', axis=1)
df = df.drop('Dt_Customer', axis=1)

df = pd.get_dummies(df, drop_first=True)
df.head()
```

## 5. Veriyi Ölçekleme

```python
# Bu bölümde clustering öncesi tüm değişkenleri aynı ölçeğe getireceğim.
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(df)
x_scaled[:5]
```

## 6. Elbow Yöntemi ile Küme Sayısı Seçme

```python
# Bu bölümde en uygun küme sayısını seçmek için inertia değerlerini inceleyeceğim.
```

```python
inertia = []
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(x_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

## 7. KMeans Modeli Kurma

```python
# Bu bölümde seçilen küme sayısı ile KMeans modelini kuracağım ve segmentleri oluşturacağım.
```

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
clusters = kmeans.fit_predict(x_scaled)

df['Cluster'] = clusters
df.head()
```

```python
silhouette = silhouette_score(x_scaled, clusters)
silhouette
```

## 8. PCA ile Görselleştirme

```python
# Bu bölümde kümeleri iki boyutta görselleştirerek segmentlerin ayrışmasını inceleyeceğim.
```

```python
pca = PCA(n_components=2)
pca_components = pca.fit_transform(x_scaled)

pca_df = pd.DataFrame(pca_components, columns=['PC1', 'PC2'])
pca_df['Cluster'] = clusters

plt.figure(figsize=(8,6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Cluster', palette='Set2')
plt.title('Customer Segments (PCA)')
plt.show()
```
